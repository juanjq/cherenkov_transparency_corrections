```python
import glob, os, subprocess

input_dl2_folder = f"/fefs/aswg/workspace/juan.jimenez/data/cherenkov_transparency_corrections/crab/dl2/"

output_dl2_radec_folder = f"/fefs/aswg/workspace/juan.jimenez/data/cherenkov_transparency_corrections/crab/dl2_radec/"
```

```python
# If output folder do not exists we create it 
os.makedirs(output_dl2_radec_folder, exist_ok=True)

# Then we find the DL2 files in the input folder
files_dl2 = glob.glob(input_dl2_folder + f"dl2_LST-1.Run?????.h5")

print(f"Number of DL2 to extract radec: {len(files_dl2)}")
```

.

```python
for i, file in enumerate(files_dl2[1:]):
    print(f"Sending jobs {i}/{len(files_dl2)} ...")

    input_file = file
    output_folder = output_dl2_radec_folder

    slurm_command = f"sbatch -p short -J dl2_add_radec --wrap='python script_add_radec.py {input_file} {output_folder}'"
    result = subprocess.run(slurm_command, shell=True, capture_output=True, text=True)
    
    print(result.stdout)
    if result.stderr:
        print(f"Error: {result.stderr}")
    
```
